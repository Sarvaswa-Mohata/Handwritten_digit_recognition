```python
  import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import torch 
import sklearn
from torchvision import datasets,transforms
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
arr_1_shuffled, arr_2_shuffled = sklearn.utils.shuffle(X_train, y_train)
arr_3_shuffled, arr_4_shuffled = sklearn.utils.shuffle(X_test, y_test)
```

```python
X_train_flattened = X_train.reshape(len(arr_1_shuffled), 28*28)

X_test_flattened = X_test.reshape(len(arr_3_shuffled), 28*28)
```

```python
# arr_1_shuffled, arr_2_shuffled = sklearn.utils.shuffle(X_train, y_train)
# # arr_1_shuffled[1].reshape(28,28)
# arr_1_shuffled[1].shape
# plt.matshow(arr_1_shuffled[1])
# arr_2_shuffled[1]
```

## Defining 1st model

```python
model_1 = keras.Sequential([
    keras.layers.Dense(150, input_shape=(784,), activation='relu'),
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_1.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model_1.fit(X_train_flattened, y_train, epochs=5)
```

## Calculating Acuuracy

```python
model_1.evaluate(X_test_flattened,y_test)
```

## Plotting confusion Matrix

```python
y_predicted = model_1.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,7))
import seaborn as sn
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
